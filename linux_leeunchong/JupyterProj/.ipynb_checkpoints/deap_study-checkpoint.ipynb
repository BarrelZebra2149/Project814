Generator algorithm implementation
this is driver code.

```python
import random

from deap import base
from deap import creator
from deap import tools
```

## Float Generator

This generator generates float.

```python
creator.create("FitnessMax", base.Fitness, weights=(1.0,))
creator.create("Individual", list, fitness=creator.FitnessMax)

IND_SIZE=10

toolbox = base.Toolbox()
toolbox.register("attr_float", random.random)
toolbox.register("individual", tools.initRepeat, creator.Individual,
                 toolbox.attr_float, n=IND_SIZE)
```

```python
ind_float = toolbox.individual()
print(ind_float)
```

## Permutation
this is permutation. it is using indices.

```python
creator.create("FitnessMin", base.Fitness, weights=(-1.0,))
creator.create("Individual", list, fitness=creator.FitnessMin)

IND_SIZE=10

toolbox = base.Toolbox()
toolbox.register("indices", random.sample, range(IND_SIZE), IND_SIZE)
toolbox.register("individual", tools.initIterate, creator.Individual,
                 toolbox.indices)
```

```python
ind_perm = toolbox.individual()
print(ind_perm)
```

## Random Number Generator?
How we can make N size of integers?

```python
creator.create("FitnessMax", base.Fitness, weights=(1.0,))
creator.create("Individual", list, fitness=creator.FitnessMax)

IND_SIZE = 20

INT_MIN, INT_MAX = 0, 9
toolbox.register("attr_int", random.randint, INT_MIN, INT_MAX)
toolbox.register("individual", tools.initRepeat, creator.Individual, 
                 toolbox.attr_int, IND_SIZE)

toolbox.register("population", tools.initRepeat, list, toolbox.individual)
```

```python
ind_int = toolbox.individual()
print(ind_int)
```
